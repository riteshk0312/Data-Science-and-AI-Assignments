# Data Splitting

```python
# stationarity check
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# define a function to check for stationary
def test_stationarity(timeseries):
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', '#Lag Used', '# of Obs'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)

    #Stationarity test
    if dfoutput[1] < 0.5 :
        print('The data is stationary')
    else:
        print('The data is not stationary')
```

```python
# call the function
```

```python
print('Augmented Dickey Fuller test')
test_stationarity(data['Rainfall_Gallicano'])
```

# Build the ARIMA model

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model  = ARIMA(data['Rainfall_Gallicano'], order = (p,d,q))
```

```python
# Train the model
```

```python
model_fit= model.fit()
```

```python
# Forecasting
forecast = model_fit.forecast(steps=72)
forecast
```

```python
plt.figure(figsize=(12,6))
plt.plot(data['Rainfall_Gallicano'],label='Original Data')
plt.plot(pd.DataFrame(forecast),label='Forecast data',color='red')
plt.legend()
plt.show()
```
